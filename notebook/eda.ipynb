```python
movies["tags"]=movies["genre"]+movies["overview"]
movies
```

```python
new_data=movies.drop(columns=["genre","overview"])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CountVectorizer()
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vectro=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vectro.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectro)
similarity
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])

for i in distance:
    print(new_data.iloc[i[0]].title)  # Use i[0] to access the row index
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])

    for i in distance:
        print(new_data.iloc[i[0]].title)  # Use i[0] to access the row index
```

```python
recommand('The Godfather')
```
